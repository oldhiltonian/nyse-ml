```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt

path = Path("CTA_-_Ridership_-_Daily_Boarding_Totals.csv")
df = pd.read_csv(path, parse_dates=['service_date'])
df.columns=['date', 'day_type', 'bus', 'rail', 'total']
df.sort_values('date').set_index('date')
df = df.drop('total', axis=1)
df = df.drop_duplicates()
```

```python
df.head()
```

```python
df.iloc[0:100].plot(grid=True, marker='.', figsize=(12,8))
```

```python
# making naive forecasts by taking the previous weeks values 
# diff_7 is the difference or error
# top plot is overlay of true and 1-week lag
# bottom plot is diff_7 or error
diff_7 = df.iloc[0:100][['bus', 'rail']].diff(7)
fig, axs = plt.subplots(2,1, sharex=True, figsize=(8,5))
df.iloc[0:100].plot(ax=axs[0], legend=False, marker='.')
df.shift(7).iloc[0:100].plot(ax=axs[0], grid=True, legend=False, linestyle=':')
diff_7.plot(ax=axs[1], grid=True, marker='.')
plt.show()
```

```python
diff_7.abs().mean() # MAE
```

```python
(diff_7/df[['bus', 'rail']].iloc[0:100]).abs().mean() #MAPE
```

```python
# Checking for yearly seasonality
# Plotting the 12-month values and their moving average
df = df.set_index('date')
period = slice("2001", "2019")
df_monthly = df.resample('M').mean() # compute mean for each month
rolling_average_12_months = df_monthly[period].rolling(window=12).mean()

fig, ax = plt.subplots(figsize=(12,5))
df_monthly[period].plot(ax=ax, marker='.')
rolling_average_12_months.plot(ax=ax, grid=True, legend=False)
plt.show()
```

```python
# Plotting the 12 month difference

df_monthly.diff(12)[period].plot(grid=True, marker='.', figsize=(12, 5))
```
